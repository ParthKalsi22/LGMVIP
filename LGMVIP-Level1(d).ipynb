### Image to Pencil Sketch with Python

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import warnings
warnings.filterwarnings("ignore")
```

```python
Image=cv2.imread("LGM.jpg")
Image
```

```python
Image_rgb = cv2.cvtColor(Image, cv2.COLOR_BGR2RGB)
plt.imshow(Image_rgb)
```

### Converting Image into Gray Scale

```python
Image_gray = cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)
Image_gray
```

```python
plt.imshow(Image_gray)
```

### Inverting the image

```python
Image_invert = cv2.bitwise_not(Image_gray)
Image_invert
```

```python
plt.imshow(Image_invert)
```

### Smoothing the image

```python
Image_smooth = cv2.GaussianBlur(Image_invert, (21, 21),sigmaX=0, sigmaY=0)
Image_smooth
```

```python
plt.imshow(Image_smooth)
```

### Final Sketch

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255- y, scale=256.0)
Final_image = dodgeV2(Image_gray, Image_smooth)
plt.imshow(Final_image)
```
